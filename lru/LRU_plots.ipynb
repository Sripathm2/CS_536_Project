```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Results.tab',sep = '\t')
```

```python

df = pd.read_csv('Results.tab',sep = '\t')
df75 = df[df['percentile']==0.75]
df751000 = df75[df75['packet_threshold']==1000]
df751500  = df75[df75['packet_threshold']==1500]
df752000  = df75[df75['packet_threshold']==2000]
df752500  = df75[df75['packet_threshold']==2500]
df753000  = df75[df75['packet_threshold']==3000]



plt.title('comparision on cache size for percentile = 0.75')
plt.xlabel('cache size')
plt.ylabel('accuracy')
x= df751000['cache_size'].to_list()
y= df751000['accuracy_lru'].to_list()
plt.plot(x,y,label='packet threshold=1000')


y1= df751500['accuracy_lru'].to_list()
plt.plot(x,y1,label='packet threshold=1500')

y2= df752000['accuracy_lru'].to_list()
plt.plot(x,y2,label='packet threshold=2000')

y3= df752500['accuracy_lru'].to_list()
plt.plot(x,y3,label='packet threshold=2500')

y4= df753000['accuracy_lru'].to_list()
plt.plot(x,y4,label='packet threshold=3000')
plt.legend()


plt.savefig('percentile0.75.png')
plt.show()

```
